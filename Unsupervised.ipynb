# Importing Libraries

```python
import pandas as pd
import numpy as np
import warnings
from sklearn.cluster import KMeans
import seaborn as sns
import matplotlib.pyplot as plt
```

# Create Synthetic Data

```python
warnings.filterwarnings("ignore") # Not to display any warnings keeping the output clean
np.random.seed(0)                 # create random data from seed value 0 to make displayed values consistent in every time you run the code
data_size = 300                   # 300 Rows of data represent no of vehicles which will be clustered based on their specifications
data ={
    'Weight': np.random.randint(1000, 3000, data_size),
    'EngineSize': np.random.uniform(1.0, 4.0, data_size),  #features of vehicles
    'Horsepower': np.random.randint(50, 300, data_size)
}
```

```python
df = pd.DataFrame(data)   # data will be read as a data frame
```

```python
x = df     #no labels to be considered
```

# Model Choice

```python
model = KMeans(n_clusters= 3, random_state=42)
model.fit(x)
```

```python
plt.scatter(df["Weight"],df["Horsepower"], c = model.labels_)
plt.xlabel("Weight")
plt.ylabel("Hoursepower")
plt.title("Clustering Groups")
plt.show()
```
